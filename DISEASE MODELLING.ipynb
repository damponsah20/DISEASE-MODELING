```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# DEFINING FUNCTIONS
def seir_model(y, t, beta, sigma, gamma):
    S, E, I, R = y
    dSdt = -beta * S * I / N  # Ensure to divide by total population N
    dEdt = beta * S * I / N - sigma * E
    dIdt = sigma * E - gamma * I
    dRdt = gamma * I
    return [dSdt, dEdt, dIdt, dRdt]

# SETTING INITIAL CONDITIONS
# Total population, N
N = 31000000  # Population of Ghana in 2020

# Initial number of infected, exposed, and recovered individuals
I0 = 200000  # Example value for HIV/AIDS
E0 = 50000  # Example value for HIV/AIDS
R0 = 0  # Example value for HIV/AIDS
# Everyone else, S0, is susceptible to infection initially
S0 = N - I0 - E0 - R0

# Contact rate, beta; incubation rate, sigma; and recovery rate, gamma
beta = 0.2  # Example value
sigma = 1/365  # Example value
gamma = 1/7300  # Example value

# A grid of time points (in days)
t = np.linspace(0, 365, 365)  # Ten years to capture the long-term dynamics of HIV/AIDS

# Initial conditions vector
y0 = S0, E0, I0, R0

# Integrate the SEIR equations over the time grid, t
ret = odeint(seir_model, y0, t, args=(beta, sigma, gamma))
S, E, I, R = ret.T

# Convert the values to millions for better readability
S = S / 1e6
E = E / 1e6
I = I / 1e6
R = R / 1e6

# Plot the data
fig = plt.figure(figsize=(10, 6))
plt.plot(t, S, 'b', alpha=0.7, linewidth=2, label='Susceptible')
plt.plot(t, E, 'y', alpha=0.7, linewidth=2, label='Exposed')
plt.plot(t, I, 'r', alpha=0.7, linewidth=2, label='Infected')
plt.plot(t, R, 'g', alpha=0.7, linewidth=2, label='Recovered')
plt.xlabel('Time (days)')
plt.ylabel('Number of people (millions)')
plt.legend()
plt.title('SEIR Model of HIV/AIDS in Ghana for 2020')
plt.show()
```
